```python
#SOAL NOMOR 1

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
```

```python
dset = pd.read_csv("C:/Users/Gilbert Ciady/Documents/Visual Studio Code Project/Semester 3/Data Mining/Tugas 1/heart_disease_uci.csv")
```

```python
#SOAL NOMOR 2.1

print("5 Baris Pertama: ")
dset.head()
```

```python
print("5 Baris Terakhir: ")
dset.tail()
```

```python
#SOAL NOMOR 2.2

print("Informasi Dataset: ")
dset_info = pd.DataFrame({
    'Data Type': dset.dtypes.values,
    'Entry': dset.count().values,
    'Column': dset.columns
})
dset_info.index = range(1, len(dset_info) + 1)
print(tabulate(dset_info, headers='keys', tablefmt='outline'))
```

```python
#SOAL NOMOR 2.3

missing_values =dset.isnull().sum().reset_index()
missing_values.columns =['Column', 'Missing Values']
missing_values.index = range(1, len(missing_values) + 1)
print("Nilai Missing atau Nilai yang Hilang: ")
print(tabulate(missing_values, headers='keys', tablefmt='outline'))
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=dset.select_dtypes(include=[np.number]))
plt.xticks(rotation=90)
plt.title('Nilai Outliers menggunakan Box Plot')
plt.show()
```

```python
#SOAL NOMOR 3.1

print("Menghitung mean, median, mode, standar deviasi, variansi, skewness, dan kurtosis")
numeric_cols = dset.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    print(f"\nStatistik untuk {col} : ")
    print(f"Mean            : {dset[col].mean()}")
    print(f"Median          : {dset[col].median()}")
    print(f"Mode            : {dset[col].mode()[0]}")
    print(f"Standar Deviasi : {dset[col].std()}")
    print(f"Variansi        : {dset[col].var()}")
    print(f"Skewness        : {dset[col].skew()}")
    print(f"Kurtosis        : {dset[col].kurt()}")
    print("-------------------------------------")
```

```python
#Menampilkan Hasil Menghitung Mean, Median, Mode, Standar Deviasi, Variansi, Skewness, dan Kurtosis dalam bentuk tabel
desc_stats = dset.describe().reset_index()
desc_stats.index = range(1, len(desc_stats) + 1)
print(tabulate(desc_stats, headers='keys', tablefmt='outline'))
```

```python
#SOAL NOMOR 3.2

numeric_dset = dset.select_dtypes(include=[np.number])
correlation_matrix = numeric_dset.corr()
correlation_matrix_reset = correlation_matrix.reset_index()
correlation_matrix_reset.index = range(1, len(correlation_matrix_reset) + 1)

print("Matriks Korelasi:")
print(tabulate(correlation_matrix_reset, headers='keys', tablefmt='outline'))

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1, linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

```python
#SOAL NOMOR 4.1

print("Histogram untuk Setiap Variabel Numerik")
dset.hist(bins=20, figsize=(15, 10))
plt.show()
```

```python
print("- Box plot untuk Setiap Variabel Numerik")
dset.plot(kind='box', subplots=True, layout=(4,4), figsize=(15,15), sharex=False, sharey=False)
plt.show()
```

```python
#SOAL NOMOR 4.2

print("Scatter Plot untuk Pasangan Variabel dengan Korelasi Tinggi")
sns.pairplot(dset, hue ='age')  
plt.show()
```

```python
max_corr_pair = numeric_dset.corr().abs().unstack().sort_values(ascending=False).drop_duplicates().index[1]

plt.figure(figsize=(8, 6))
sns.regplot(x=numeric_dset[max_corr_pair[0]], y=numeric_dset[max_corr_pair[1]], order=2, scatter_kws={'s': 50}, line_kws={"color":"red"})
plt.title(f'Scatter Plot antara {max_corr_pair[0]} dan {max_corr_pair[1]} dengan Garis Regresi Polinomial (Korelasi: {numeric_dset.corr().loc[max_corr_pair[0], max_corr_pair[1]]:.2f})')
plt.xlabel(max_corr_pair[0])
plt.ylabel(max_corr_pair[1])
plt.show()
```
